```python
# Datos de entrenamiento
import pandas as pd

train = pd.read_csv("/content/EXIST2021_training.tsv", delimiter='\t', usecols=['language','text', 'task2'], quoting=3, error_bad_lines=False)
test = pd.read_csv("/content/EXIST2021_test_labeled.tsv", delimiter='\t', usecols=['language','text', 'task2'], quoting=3, error_bad_lines=False)
```

<font color="brown"><h1>**Visualización de características**</h1></font>

```python
# Feature Engineering
train_ca = train.copy()
train_ca['longitud_text'] = train_ca['text'].str.len()
train_ca['num_hashtags'] = train_ca['text'].str.count('#')
train_ca['num_exclam_signos'] = train_ca['text'].str.count('\!')
train_ca['num_interrog_signos'] = train_ca['text'].str.count('\?')
train_ca['total_tags'] = train_ca['text'].str.count('@')
train_ca['num_punt_signos'] = train_ca['text'].str.count('[.,:;]')
train_ca['num_interrog_signos'] = train_ca['text'].str.count('[*&$%]')
train_ca['num_palabras'] = train_ca['text'].apply(lambda x: len(x.split()))
train_ca['num_http'] = train_ca['text'].str.count(r'http')
train_ca['num_https'] = train_ca['text'].str.count(r'https')
train_ca.head(25)
```

```python
# Feature Engineering
test_ca = test.copy()
test_ca['longitud_text'] = test_ca['text'].str.len()
test_ca['num_hashtags'] = test_ca['text'].str.count('#')
test_ca['num_exclam_signos'] = test_ca['text'].str.count('\!')
test_ca['num_interrog_signos'] = test_ca['text'].str.count('\?')
test_ca['total_tags'] = test_ca['text'].str.count('@')
test_ca['num_punt_signos'] = test_ca['text'].str.count('[.,:;]')
test_ca['num_interrog_signos'] = test_ca['text'].str.count('[*&$%]')
test_ca['num_palabras'] = test_ca['text'].apply(lambda x: len(x.split()))
test_ca['num_http'] = test_ca['text'].str.count(r'http')
test_ca['num_https'] = test_ca['text'].str.count(r'https')
test_ca.head(25)
```

```python
train_ca.to_csv('train_ca.csv', index=False)
test_ca.to_csv('test_ca.csv', index=False)
```
